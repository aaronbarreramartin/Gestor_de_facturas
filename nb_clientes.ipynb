### En un Notebook como este podríamos analizar los servicios de cada cliente, cuál genera más beneficios, y cualquier característica de interés.

```python
from pathlib import Path
import pandas as pd

csvs = list(Path.glob(Path.cwd(), 'csvs/*.csv'))
dfs = [(pd.read_csv(f, parse_dates=['Fecha']), f.name) for f in csvs]
```

```python
for df, nombre in dfs:
    print(nombre + ':')
    display(df)
```

```python
df_completo = pd.DataFrame()
for df, nombre in dfs:
    df['Cliente'] = nombre.split('-')[0]
    df_completo = pd.concat([df_completo, df])
df_completo
```

```python
df_completo.groupby('Concepto')[['Importe']].sum()
```

```python
df_completo.groupby('Cliente')[['Importe']].sum()
```
